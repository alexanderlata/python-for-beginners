# Функции в Python

## 1. Что такое функция

<div class="alert alert-block alert-success">

**Функция** в Python — это именованный блок кода, который выполняет определённую задачу и может быть вызван многократно. Функции помогают структурировать программу, делая её более читаемой и поддерживаемой.

Например, в игре «Тетрис» обычно есть отдельные функции для подсчёта очков, рисования игрового поля, движения фигурки и т. д.

</div>

**Ключевые особенности функций:**

* Функция выполняется только тогда, когда её вызывают.
* В функцию можно передавать данные — через параметры.
* Функция может возвращать результат своей работы.
* Результат работы функции можно использовать в другой функции или в основной программе.

**Использование функций позволяет:**

* исключить дублирование кода;
* разбить большую программу на небольшие, легко проверяемые части;
* ограничить область видимости переменных рамками функции;
* упростить отладку и поддержку программы.

## 2. Параметры и аргументы

<div class="alert alert-block alert-success">

**Параметры** — это переменные, которые указываются при *объявлении* функции.

**Аргументы** — фактические значения, которые передаются в функцию при её *вызове*.

</div>

Эти понятия часто путают, но различие простое:

```python
def greet(name):          # name — параметр
    print(f'Привет, {name}')

greet('Мария')            # 'Мария' — аргумент
```

В Python есть **встроенные** функции (`print()`, `input()`, `len()`, `map()`, `zip()` и т. д.) и **пользовательские** функции — те, которые пишет сам программист.

## 3. Синтаксис объявления функции

Для объявления функции используется ключевое слово `def` (от англ. *define* — определять).

```python
def <имя_функции>([<параметры>]):
    """строка документации (docstring)"""
    <тело функции>
    [return <возвращаемое значение>]
```

* В квадратных скобках указаны **необязательные** элементы.
* Параметры перечисляются через запятую. Если их нет — круглые скобки всё равно обязательны.
* После списка параметров ставится двоеточие.
* **Тело функции выделяется отступом в 4 пробела** (как у условий и циклов).
* Необязательный **docstring** описывает, что делает функция.

<div class="alert alert-block alert-warning">

Правила записи **имени функции** такие же, как у переменных: строчные буквы английского алфавита, цифры и знак подчёркивания. Имя должно начинаться с буквы или `_` и быть осмысленным (`calc_price`, а не `f1`).

</div>

## 6. Возврат значения: оператор `return`

Оператор `return` завершает работу функции и возвращает значение в то место, откуда функция была вызвана.

<div class="alert alert-block alert-warning">

**Важно:** функция в Python **всегда** возвращает значение. Если в функции нет `return` (или `return` стоит без значения), функция возвращает `None` — специальное значение, которое означает «ничего».

</div>

`None` возвращает, например, встроенная функция `print()`:

````python
# Внутренний print выводит строку и возвращает None
# Внешний print получает None и выводит его
print(print('Эту строку выведет внутренний print'))
````

### 6.1. Функция, возвращающая результат

````python
def calc_product(a, b, c):
    return a * b * c

result = calc_product(5, 6, 3)
print(result)
````

### 6.2. `print()` vs `return`

Новички часто путают `print()` и `return`. Разница принципиальная:

* `return` **возвращает** значение вызывающему коду — его можно сохранить в переменную или использовать в выражении.
* `print()` **выводит** значение на экран — но из функции ничего не возвращает.

````python
def double_with_print(x):
    print(x * 2)       # только вывод на экран

def double_with_return(x):
    return x * 2       # возвращает значение

# Используем обе функции в выражении:
a = double_with_print(5)    # выведет 10, но a = None
b = double_with_return(5)   # ничего не выведет, но b = 10

print(f'a = {a}')
print(f'b = {b}')
print(f'Можем использовать return-результат: {double_with_return(5) + 100}')
````

<div class="alert alert-block alert-success">

Правило: если результат работы функции нужен для дальнейших вычислений — используй `return`. `print()` внутри функции оставляй только там, где функция должна именно *показать* что-то пользователю.

</div>

## 7. Несколько операторов `return` в функции

В теле функции может быть несколько `return`. Выполнится **первый, до которого дойдёт управление**, — и на этом функция завершится.

Напишем функцию, которая проверяет, что все числа в списке — чётные. Сначала — с использованием переменной-флага:

````python
def only_even_v1(numbers):
    result = True
    for x in numbers:
        if x % 2 != 0:
            result = False
            break
    return result

print(only_even_v1([2, 4, 6]))
print(only_even_v1([1, 2, 3]))
````

Этот же код можно упростить, если использовать несколько точек возврата:

````python
def only_even_v2(numbers):
    for x in numbers:
        if x % 2 != 0:
            return False
    return True

print(only_even_v2([2, 4, 6]))
print(only_even_v2([1, 2, 3]))
````

### 7.1. Возврат нескольких значений

Функция может вернуть сразу несколько значений — их перечисляют через запятую после `return`. Технически возвращается **кортеж**.

````python
def calculations(a, b):
    summa = a + b
    diff = a - b
    mul = a * b
    div = a / b
    return summa, diff, mul, div

# Результат можно распаковать сразу в несколько переменных:
s, d, m, q = calculations(10, 3)
print(f'Сумма: {s}')
print(f'Разность: {d}')
print(f'Произведение: {m}')
print(f'Деление: {q:.2f}')
````

<div class="alert alert-block alert-warning">

**Хороший стиль:** функция должна возвращать значения **одного типа** при любом ходе выполнения. Например, если функция иногда возвращает `int`, а иногда `None` — это нормально (и часто удобно). А вот если функция то возвращает `bool`, то кортеж `(bool, int)` — это плохая практика: вызывающий код вынужден проверять тип возврата.

Если нужно вернуть несколько значений, возвращайте их всегда, используя при необходимости «пустые» значения-заглушки:

</div>

````python
def first_odd_index(numbers):
    """Возвращает (True, -1) если все чётные, иначе (False, индекс первого нечётного)."""
    for i, x in enumerate(numbers):
        if x % 2 != 0:
            return False, i
    return True, -1

print(first_odd_index([2, 4, 6]))
print(first_odd_index([1, 2, 3]))
````

## 8. Заглушка `pass`

Тело функции не может быть пустым — попытка написать `def` без тела приведёт к `IndentationError`:

```python
def my_function():   # IndentationError
# (следующая строка не имеет отступа)
```

Если функцию нужно объявить, но пока нечего в неё писать (например, планируем дописать позже), используют оператор `pass` — он ничего не делает и служит заглушкой:

````python
def not_implemented_yet():
    pass

# Функцию можно вызвать, она просто ничего не сделает и вернёт None
result = not_implemented_yet()
print(result)
````

## 9. Область видимости переменных

**Область видимости** — это то, где в программе переменная существует и доступна.

В Python есть несколько областей видимости, которые Python проверяет в строгом порядке — правило **LEGB**:

* **L** — **Local**: локальная область текущей функции
* **E** — **Enclosing**: область внешней функции (если функция вложена в другую)
* **G** — **Global**: область модуля (основной код программы)
* **B** — **Built-in**: встроенные имена Python (`print`, `len`, `range`, ...)

Когда Python встречает имя переменной, он ищет её именно в таком порядке: сначала в локальной области, потом во внешней функции, потом в глобальной, потом среди встроенных имён.

### 9.1. Локальная область

Параметры и переменные, объявленные внутри функции, доступны **только внутри неё**:

````python
def demo(numbers):
    total = sum(numbers)
    return total

demo([1, 2, 3])
````

````python
# Попытка обратиться к параметру функции извне — ошибка
print(numbers)  # NameError: name 'numbers' is not defined
````

````python
# То же самое с локальной переменной
print(total)    # NameError
````

### 9.2. Глобальная область

Переменные, объявленные в основном коде программы (вне функций), находятся в глобальной области видимости. К ним можно обращаться **на чтение** из любой функции.

````python
# Глобальная переменная MAX_DISCOUNT — читается внутри функции
MAX_DISCOUNT = 90

def is_valid_discount(d):
    return 0 <= d <= MAX_DISCOUNT

print(is_valid_discount(50))
print(is_valid_discount(95))
````

<div class="alert alert-block alert-warning">

**Хороший стиль:** старайтесь передавать все нужные функции данные через её параметры, а не тянуть из глобальной области. Исключение — настоящие константы (как `MAX_DISCOUNT` выше). Код, в котором функции зависят от многих глобальных переменных, очень сложно отлаживать.

</div>

### 9.3. Изменяемые и неизменяемые типы в глобальной области

Если глобальная переменная — **изменяемого** типа (список, словарь, множество), её содержимое можно изменить изнутри функции:

````python
def remove_last():
    del sample[-1]   # меняем сам список

sample = [1, 2, 3]
remove_last()
print(sample)
````

Но если внутри функции сделать **присваивание** переменной с тем же именем, что и глобальная, Python создаст новую **локальную** переменную, а глобальная не изменится:

````python
def try_replace():
    sample = [4, 5, 6]   # это ЛОКАЛЬНАЯ переменная функции

sample = [1, 2, 3]
try_replace()
print(sample)   # глобальный sample не изменился
````

Это же правило работает и для параметра функции — он тоже локальная переменная:

````python
def modify_via_assign(list_arg):
    list_arg = [1, 2, 3, 4]   # создаёт новый локальный список

def modify_in_place(list_arg):
    list_arg += [4]           # меняет исходный список

s1 = [1, 2, 3]
s2 = [1, 2, 3]
modify_via_assign(s1)
modify_in_place(s2)
print('s1:', s1)   # не изменился
print('s2:', s2)   # изменился
````

### 9.4. Ключевое слово `global`

Если в функции нужно **перезаписать** глобальную переменную (а не просто прочитать), пишут `global <имя>`:

````python
call_count = 0

def inc():
    global call_count
    call_count += 1
    print(f'Функция вызвана {call_count} раз(а).')

inc()
inc()
inc()
````

<div class="alert alert-block alert-warning">

**Использование `global` почти всегда можно (и нужно) заменить** передачей значения через параметр и возвратом через `return`. Код с глобальным состоянием тяжело тестировать и отлаживать.

</div>

Тот же счётчик без `global`:

````python
def inc(count):
    count += 1
    print(f'Функция вызвана {count} раз(а).')
    return count

call_count = 0
call_count = inc(call_count)
call_count = inc(call_count)
call_count = inc(call_count)
````

## 10. Позиционные и именованные аргументы. Значения по умолчанию

### 10.1. Позиционные аргументы

Когда мы просто перечисляем аргументы при вызове, Python сопоставляет их с параметрами **слева направо** — такие аргументы называются **позиционными**.

````python
def final_price(price, discount):
    return price - price * discount / 100

print(final_price(1000, 5))
````

### 10.2. Значения по умолчанию

Параметру можно задать значение по умолчанию. Если при вызове аргумент не передан — используется это значение.

````python
def final_price(price, discount=1):
    return price - price * discount / 100

print(final_price(1000, 5))   # с явной скидкой 5%
print(final_price(1000))      # используется значение по умолчанию: 1%
````

### 10.3. Ловушка: изменяемое значение по умолчанию

<div class="alert alert-block alert-warning">

**Значение по умолчанию вычисляется один раз — в момент определения функции**, а не при каждом её вызове. Поэтому, если в качестве значения по умолчанию использовать изменяемый объект (список, словарь, множество), он будет **один и тот же** для всех вызовов!

</div>

````python
def add_value_BAD(x, list_arg=[]):
    list_arg += [x]
    return list_arg

print(add_value_BAD(0))          # ожидаем [0]
print(add_value_BAD(0, [1, 2, 3]))  # [1, 2, 3, 0]
print(add_value_BAD(1))          # ожидаем [1], получим [0, 1]!
````

Правильный приём — использовать `None` как значение по умолчанию и создавать новый объект внутри функции:

````python
def add_value(x, list_arg=None):
    if list_arg is None:
        list_arg = []
    list_arg += [x]
    return list_arg

print(add_value(0))
print(add_value(0, [1, 2, 3]))
print(add_value(1))
````

### 10.4. Именованные аргументы

Аргумент можно передать по имени параметра — в виде `имя=значение`. Такой аргумент называется **именованным** (keyword argument).

Это удобно, когда параметров много и порядок сложно запомнить.

````python
def final_price(price, discount=1):
    return price - price * discount / 100

print(final_price(1000, discount=5))           # смешанно
print(final_price(discount=10, price=1000))    # можно в любом порядке
````

<div class="alert alert-block alert-warning">

**Порядок при вызове функции:** сначала идут позиционные аргументы, затем именованные. Обратный порядок — синтаксическая ошибка.

</div>

## 11. Функции с переменным числом аргументов

Иногда нужно, чтобы функция принимала произвольное количество аргументов (как встроенная `print()` — ей можно передать любое число строк).

### 11.1. `*args` — переменное число позиционных аргументов

Если при объявлении функции перед именем параметра поставить `*`, все «лишние» позиционные аргументы будут собраны в **кортеж** с этим именем. По соглашению его называют `args` (сокращение от *arguments*).

````python
def final_price(*prices, discount=1):
    return [p - p * discount / 100 for p in prices]

print(final_price(100, 200, 300, discount=5))
````

````python
def min_max(*args):
    print(f'Минимум: {min(args)}, максимум: {max(args)}')

min_max(1, 4, 5, 2, -5, 0, 12, 11)
````

Внутри функции с `args` можно работать как с обычным кортежем:

````python
def first_and_last(*args):
    print(f'Первый: {args[0]}, последний: {args[-1]}')

first_and_last('яблоко', 'виноград', 'апельсин', 'арбуз')
````

Имя `args` — лишь соглашение, можно выбрать любое осмысленное:

````python
def show_cities(*cities):
    print(f'Всего городов: {len(cities)}, первый: {cities[0]}')

show_cities('Тюмень', 'Москва', 'Орёл', 'Новгород')
````

### 11.2. `**kwargs` — переменное число именованных аргументов

Два символа `**` перед именем параметра «собирают» все именованные аргументы, которые не соответствуют явно объявленным параметрам, в **словарь**. По соглашению его называют `kwargs` (от *keyword arguments*).

````python
def show_countries(**countries):
    print(f'Получено стран: {len(countries)}')
    for name, density in countries.items():
        print(f'  {name}: {density} чел/км²')

show_countries(Мальта=1432, Дания=128, Монако=18679, Индия=357)
````

### 11.3. Порядок параметров в объявлении функции

Порядок следования разных видов параметров в `def` строгий:

1. Обычные позиционные параметры
2. `*args`
3. Параметры со значениями по умолчанию / только-именованные
4. `**kwargs`

````python
def demo(x, y, *args, kx=15, ky=15, **kwargs):
    print(f'x={x}, y={y}')
    print(f'args={args}')
    print(f'kx={kx}, ky={ky}')
    print(f'kwargs={kwargs}')

demo(7, 8, 0, 3, 4, север=15, запад=25, юг=10)
````

## 12. Аннотация типов

**Аннотация функции** (*function annotation*) позволяет подсказать тип параметров и возвращаемого значения.

* Тип параметра указывается после двоеточия: `x: float`
* Тип возвращаемого значения — после стрелки `->` перед двоеточием в заголовке.

````python
# x — ожидается float со значением по умолчанию 3.5, возвращается int
def to_int(x: float = 3.5) -> int:
    return int(x)

to_int()
````

Аннотации доступны через атрибут `__annotations__`:

````python
to_int.__annotations__
````

<div class="alert alert-block alert-warning">

**Аннотации не обязательны и не влияют на работу функции.** Python не проверяет, действительно ли в функцию передали значение указанного типа. Аннотации нужны:

* для документирования и читаемости кода;
* для проверки типов внешними инструментами (например, `mypy`);
* для автодополнения и подсказок в IDE.

</div>

Покажем, что аннотации реально ничего не проверяют:

````python
def to_int(x: float) -> int:
    return float(x)   # несмотря на аннотацию -> int, возвращаем float

result = to_int(3)   # передаём int вместо float — тоже никаких ошибок
print(result, type(result))
````

## 13. Функции высших порядков

**Функция высшего порядка** — это функция, которая либо **принимает** другую функцию как аргумент, либо **возвращает** функцию как результат.

Чтобы передать функцию как аргумент, укажите её имя **без круглых скобок** — круглые скобки означают вызов, а нам нужно передать саму функцию.

### 13.1. `filter()` — отбор элементов по условию

`filter(func, iterable)` возвращает итератор с теми элементами `iterable`, для которых `func(элемент)` возвращает `True`.

````python
def is_positive(x):
    return x > 0

result = filter(is_positive, [-1, 5, 6, -10, 0])
print(list(result))
````

В качестве функции-критерия можно использовать любую функцию, возвращающую `bool`. Например, метод `str.isalpha` для отбора только буквенных символов:

````python
result = filter(str.isalpha, '123ABcd()')
print(''.join(result))
````

### 13.2. `map()` — применение функции к каждому элементу

`map(func, iterable)` возвращает итератор, в котором каждый элемент — это результат применения `func` к элементу `iterable`.

````python
def square(x):
    return x ** 2

result = map(square, range(5))
print(list(result))
````

Частый приём — прочитать список чисел со стандартного ввода:

````python
# numbers = list(map(int, input('Введите числа через пробел: ').split()))
# print(numbers)

# Для демонстрации без input:
raw = '10 20 30 40 50'
numbers = list(map(int, raw.split()))
print(numbers)
````

## 14. Лямбда-функции (анонимные функции)

**Лямбда-функция** — это функция, определяемая одним **выражением** с помощью ключевого слова `lambda`. Обычно ей не дают имя, поэтому лямбда-функции называют **анонимными**.

Термин «лямбда» пришёл из формальной системы $\lambda$-исчисления.

<div class="alert alert-block alert-warning">

**Синтаксис:**

```python
lambda аргументы: выражение
```

Лямбда-функция может иметь любое число аргументов, но **только одно выражение**. Слово `return` не пишется — результат вычисления выражения возвращается автоматически.

</div>

### 14.1. Лямбда vs обычная функция

Обычная функция для вычисления дискриминанта:

````python
def discr(a, b, c):
    return b ** 2 - 4 * a * c

print(discr(1, 5, 6))
````

То же самое лямбда-функцией:

````python
discr_lambda = lambda a, b, c: b ** 2 - 4 * a * c
print(discr_lambda(1, 5, 6))
````

<div class="alert alert-block alert-warning">

**Важно:** присваивать лямбда-функцию имени (как в примере выше: `discr_lambda = lambda ...`) — это плохая практика по PEP 8. Если функции нужно имя — используйте обычный `def`. Лямбды создают **для передачи в другие функции**, не для хранения в переменных.

</div>

### 14.2. Типичное применение: лямбда внутри `sorted`, `map`, `filter`

Раньше мы использовали `filter` и `map` с обычными функциями. Часто функция-критерий состоит из одной строки — для таких случаев удобна лямбда:

````python
result = filter(lambda x: x > 0, [-1, 5, 6, -10, 0])
print(list(result))
````

````python
result = map(lambda x: x ** 2, range(5))
print(list(result))
````

### 14.3. Лямбда и сортировка

У `sorted()` есть именованный аргумент `key` — функция, значение которой используется как критерий сортировки. Лямбда здесь особенно уместна.

````python
# Сортировка строк по длине
lines = ['abcd', 'ab', 'abc', 'abcdef']
print(sorted(lines, key=lambda s: len(s)))
````

Можно задать несколько критериев — вернуть кортеж. Сортировка идёт сначала по первому элементу кортежа, при равенстве — по второму и т. д.

````python
# Сначала по длине, при равной длине — по алфавиту
lines = ['abcd', 'ab', 'ba', 'acde']
print(sorted(lines, key=lambda s: (len(s), s)))
````

Чтобы изменить направление сортировки по числовому критерию, поставьте минус:

````python
# По убыванию длины, при равной длине — по алфавиту (по возрастанию)
lines = ['abcd', 'ab', 'ba', 'acde']
print(sorted(lines, key=lambda s: (-len(s), s)))
````

Сортировка словаря по длине ключа:

````python
mydict = {'слива': 5, 'папайя': 6, 'лук': 56, 'маракуйя': 78, 'ежевика': 45}

print('До сортировки:   ', mydict)
sorted_items = sorted(mydict.items(), key=lambda item: len(item[0]))
print('После сортировки:', dict(sorted_items))
````

Сортировка списка словарей по значению одного из ключей:

````python
fruit = [
    {'название': 'виноград', 'цена': 230, 'количество': 412},
    {'название': 'манго',    'цена': 350, 'количество': 21},
    {'название': 'бананы',   'цена': 70,  'количество': 234},
    {'название': 'яблоки',   'цена': 66,  'количество': 213},
]

for item in sorted(fruit, key=lambda x: x['цена']):
    print(item)
````

### 14.4. Условное выражение в лямбде

В лямбде можно использовать условный оператор в форме выражения: `A if условие else B`.

````python
vowels = 'аиеёоуыэюя'
st = 'яжертыуиопшщасдфгчйклзхцвбнм'
result = list(map(lambda c: 'гласная' if c in vowels else 'согласная', st))
print(result[:5])   # покажем только первые 5 элементов
````

````python
# Отбор чётных чисел, превосходящих 6
numbers = [7, 8, 6, 9, 4, -6, 2, 0, -3, -12, -5, -2, 12, 77, 32]
print(list(filter(lambda x: x > 6 and x % 2 == 0, numbers)))
````

### 14.5. Ограничения лямбды

* **Только одно выражение.** Нельзя использовать операторы (`return`, `if`/`else` как команду, `for`, `while`, `pass`, `assert`, `raise` и т. д.). Можно только выражения — в том числе условное выражение `A if cond else B` и включения (comprehensions).
* Если кажется, что в лямбде нужны несколько инструкций, — это сигнал, что пора писать обычную функцию через `def`.

### 14.6. Лямбда как возвращаемое значение функции

Функция может **вернуть** другую функцию (в том числе лямбду). Это пример функции высшего порядка — не принимающей, а возвращающей функцию.

````python
def make_multiplier(n):
    """Возвращает функцию, которая умножает аргумент на n."""
    return lambda a: a * n

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(15))
print(triple(15))
````
