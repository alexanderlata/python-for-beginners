# Условный оператор в Python

Условный оператор используется для выполнения различных действий в зависимости от выполнения условия. В Python основным условным оператором является `if`, который проверяет выражение и, если оно истинно, выполняет соответствующий блок кода.

Благодаря условным операторам программа может **принимать решения** — выбирать, какой из нескольких вариантов кода выполнить.

## Операторы сравнения

Прежде чем разбирать условный оператор, нужно познакомиться с **операторами сравнения** — они используются внутри условий и возвращают результат `True` (истина) или `False` (ложь).

| Оператор | Значение           | Пример    | Результат |
|----------|--------------------|-----------|----------|
| `==`     | равно              | `5 == 5`  | `True`   |
| `!=`     | не равно           | `5 != 3`  | `True`   |
| `<`      | меньше             | `3 < 7`   | `True`   |
| `>`      | больше             | `7 > 3`   | `True`   |
| `<=`     | меньше или равно   | `5 <= 5`  | `True`   |
| `>=`     | больше или равно   | `6 >= 10` | `False`  |
| `in`     | входит в коллекцию | `2 in [1, 2, 3]` | `True` |
| `not in` | не входит в коллекцию | `5 not in [1, 2, 3]` | `True` |

````python
# Попробуем операторы сравнения
print(5 == 5)       # True
print(5 != 3)       # True
print(3 < 7)        # True
print(6 >= 10)      # False
print(2 in [1, 2, 3])      # True
print(5 not in [1, 2, 3])  # True
````

## Логические операторы

Логические операторы позволяют **комбинировать несколько условий**:

| Оператор | Значение        | Описание                                              |
|----------|-----------------|-------------------------------------------------------|
| `and`    | логическое «И»  | `True`, если **оба** условия истинны                  |
| `or`     | логическое «ИЛИ»| `True`, если **хотя бы одно** условие истинно         |
| `not`    | логическое «НЕ» | Инвертирует значение: `True` → `False`, `False` → `True` |

````python
# Оператор and — истина, если оба условия выполнены
age = 20
has_ticket = True
print(age >= 18 and has_ticket)  # True
````

````python
# Оператор or — истина, если хотя бы одно условие выполнено
is_weekend = False
is_holiday = True
print(is_weekend or is_holiday)  # True
````

````python
# Оператор not — инвертирует условие
is_raining = False
print(not is_raining)  # True — дождя нет, значит можно идти гулять
````

### Краткая форма — `if`

```python
if условие:
    # блок кода — выполняется, если условие истинно
```

Если выражение в `условие` истинно (`True`), выполняется блок кода под `if`. Иначе этот блок **пропускается**.

````python
x = 10
if x > 5:
    print(f"{x} больше 5")
````

````python
# Пример с вводом данных от пользователя
temperature = int(input("Введите температуру на улице (°C): "))
if temperature < 0:
    print("На улице мороз, оденьтесь теплее!")
````

### Полная форма — `if-else`

```python
if условие:
    # блок кода, если условие истинно
else:
    # блок кода, если условие ложно
```

Блок `else` выполняется тогда, когда условие в `if` оказалось ложным (`False`).  
Таким образом, **всегда выполняется ровно один** из двух блоков.

````python
x = 3
if x % 2 == 0:
    print(f"{x} — чётное число")
else:
    print(f"{x} — нечётное число")
````

````python
# Проверка доступа по возрасту
age = int(input("Введите ваш возраст: "))
if age >= 18:
    print("Добро пожаловать!")
else:
    print("Вам нет 18 лет. Доступ ограничен.")
````

### Множественное ветвление — `if-elif-else`

```python
if условие1:
    # блок кода, если условие1 истинно
elif условие2:
    # блок кода, если условие1 ложно, а условие2 истинно
elif условие3:
    # блок кода, если условие1 и условие2 ложны, а условие3 истинно
else:
    # блок кода, если ни одно из условий не истинно
```

Условия проверяются **по порядку сверху вниз**. Как только одно из них оказывается истинным — выполняется соответствующий блок, остальные проверки пропускаются.

````python
# Пример 1: Определение оценки по баллам
grade = int(input("Введите балл (0-100): "))

if grade >= 90:
    letter = "Отлично"
elif grade >= 75:
    letter = "Хорошо"
elif grade >= 60:
    letter = "Удовлетворительно"
else:
    letter = "Неудовлетворительно"

print(f"Балл: {grade} — {letter}")
````

````python
# Пример 2: Определение времени года по номеру месяца
month = int(input("Введите номер месяца (1-12): "))

if 3 <= month <= 5:
    season = "весна"
elif 6 <= month <= 8:
    season = "лето"
elif 9 <= month <= 11:
    season = "осень"
else:
    season = "зима"

print(f"Месяц {month} — это {season}")
````

### Вложенные условные операторы

Операторы `if` можно располагать **внутри других операторов `if`** — это называется вложенными условиями. Они позволяют проверять более сложные ситуации.

<div class="alert alert-success">

**Совет:** часто вложенные условия можно заменить на одно условие с оператором `and`. Это делает код проще и читабельнее.

</div>

````python
# Вариант с вложенными условиями
x = 12
if x > 0:
    if x % 2 == 0:
        print("Положительное чётное число")
    else:
        print("Положительное нечётное число")
else:
    print("Число не положительное")
````

````python
# Тот же результат через оператор and — код короче и понятнее
x = 12
if x > 0 and x % 2 == 0:
    print("Положительное чётное число")
elif x > 0 and x % 2 != 0:
    print("Положительное нечётное число")
else:
    print("Число не положительное")
````

````python
# Ещё пример: проверка доступа по логину и паролю
correct_login = "admin"
correct_password = "1234"

login = input("Введите логин: ")
if login == correct_login:
    password = input("Введите пароль: ")
    if password == correct_password:
        print("Вход выполнен успешно!")
    else:
        print("Неверный пароль")
else:
    print("Неверный логин")
````

### Проверка истинности значений

В Python любое значение можно использовать в условии напрямую — без явного сравнения.  
Следующие значения считаются **ложными** (`False`):

| Тип        | Ложное значение |
|------------|-----------------|
| `bool`     | `False`         |
| `NoneType` | `None`          |
| `int`      | `0`             |
| `float`    | `0.0`           |
| `str`      | `''`            |
| `list`     | `[]`            |
| `dict`     | `{}`            |
| `tuple`    | `()`            |
| `set`      | `set()`         |

Все остальные значения считаются **истинными** (`True`).

````python
# Проверка строки на пустоту
name = input("Введите ваше имя: ")
if name:
    print(f"Привет, {name}!")
else:
    print("Вы не ввели имя")
````

````python
# Проверка списка на пустоту
cart = []  # пустая корзина
if cart:
    print(f"В корзине {len(cart)} товаров")
else:
    print("Корзина пуста")
````

````python
# Проверка словаря на пустоту
student_grades = {}  # пустой словарь
if student_grades:
    print("Оценки найдены")
else:
    print("Оценок пока нет")
````

### Тернарный оператор

<div class="alert alert-success">

- Это **сокращённая запись** конструкции `if-else` в одну строку.  
- Используйте её только тогда, когда хорошо освоили обычный `if-else` — и только для простых условий.

</div>

**Синтаксис:**
```python
результат = значение1 if условие else значение2
```

Читается так: «присвоить `значение1`, **если** `условие` истинно, **иначе** `значение2`».

````python
# Обычный if-else
age = 20
if age >= 18:
    status = "взрослый"
else:
    status = "ребёнок"
print(status)
````

````python
# То же самое через тернарный оператор
age = 20
status = "взрослый" if age >= 18 else "ребёнок"
print(status)
````

````python
# Ещё пример: проверка окончания строки
text = input("Введите предложение: ")
result = "оканчивается точкой" if text.endswith(".") else "не оканчивается точкой"
print(f"Строка {result}")
````

## Примеры использования условного оператора с разными типами данных

Вы уже знаете строки, списки, кортежи, множества и словари — давайте посмотрим, как условный оператор работает с каждым из этих типов.

### С числами

````python
number = int(input("Введите число: "))
if number > 0:
    print("Положительное число")
elif number < 0:
    print("Отрицательное число")
else:
    print("Ноль")
````

### Со строками

````python
text = input("Введите слово или фразу: ")

if text.startswith("П"):
    print("Строка начинается с буквы П")

if len(text) > 10:
    print("Это длинная строка (более 10 символов)")
else:
    print("Строка короткая (10 символов или меньше)")

if text.isupper():
    print("Строка написана ЗАГЛАВНЫМИ буквами")
````

### Со списками

````python
fruits = ["яблоко", "банан", "апельсин"]
fruit = input("Введите название фрукта: ")

if fruit in fruits:
    print(f"{fruit} есть в списке")
else:
    print(f"{fruit} не найден в списке")

if len(fruits) > 0:
    print(f"В списке {len(fruits)} фрукта(-ов)")
````

### Со словарями

````python
grades = {"Алиса": 90, "Иван": 74, "Мария": 55}
student = input("Введите имя студента: ")

if student in grades:
    score = grades[student]
    if score >= 75:
        result = "зачёт"
    else:
        result = "незачёт"
    print(f"{student}: {score} баллов — {result}")
else:
    print(f"Студент {student} не найден в базе")
````

### С множествами

````python
allowed_cities = {"Москва", "Санкт-Петербург", "Казань"}
city = input("Введите название города: ")

if city in allowed_cities:
    print(f"{city} входит в список доступных городов")
else:
    print(f"{city} не входит в список доступных городов")
````

### С кортежами

````python
valid_statuses = ("active", "pending", "verified")
status = input("Введите статус пользователя: ")

if status in valid_statuses:
    print(f"Статус '{status}' является допустимым")
else:
    print(f"Статус '{status}' недопустим")
````
