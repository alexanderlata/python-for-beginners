# Включения в Python

## 1. Введение во включения

**Включение** (англ. *comprehension*) — это компактная синтаксическая конструкция, которая позволяет **создать новую коллекцию** (список, множество или словарь) на основе уже существующей последовательности в одну строку.

Идея очень проста: вместо того чтобы писать цикл `for` с `append()` (или с добавлением в множество/словарь), мы описываем сразу **как должна выглядеть готовая коллекция** — и Python собирает её за нас.

В Python есть три основных вида включений:

| Вид | Скобки | Что создаёт |
|---|---|---|
| Списковое включение | `[ ... ]` | `list` |
| Множественное включение | `{ ... }` | `set` |
| Словарное включение | `{ключ: значение ...}` | `dict` |

<div class="alert alert-success">

**Зачем это нужно?** Включения делают код короче и, если пользоваться ими с умом, гораздо читаемее. Такой стиль записи считается «питоничным» (*Pythonic*) — то есть наиболее естественным для Python.

</div>

## 2. Списковое включение (list comprehension)

__Базовый синтаксис__
```python
[выражение for переменная in итерируемый_объект]
```

Читается это так: «собери в список *выражение* для каждой *переменной* из *итерируемого объекта*».

Работает включение абсолютно так же, как цикл `for` — просто записано компактнее.

### Сравнение: обычный цикл и списковое включение

Допустим, нужно получить список квадратов чисел от 0 до 4.

````python
# Вариант 1: обычный цикл for
squares = []
for x in range(5):
    squares.append(x ** 2)

print(squares)   # [0, 1, 4, 9, 16]
````

````python
# Вариант 2: списковое включение — тот же результат в одну строку
squares = [x ** 2 for x in range(5)]

print(squares)   # [0, 1, 4, 9, 16]
````

__Объяснение:__ Оба варианта делают одно и то же:
1. Создают пустой список.
2. Перебирают числа из `range(5)`.
3. Для каждого числа вычисляют `x ** 2` и добавляют результат в список.

Разница только в форме записи: включение убирает явный `append()` и заголовок цикла помещает внутрь квадратных скобок.

### Ещё примеры простых включений

````python
# Список длин слов
words = ['Python', 'МГИМО', 'включение', 'список']
lengths = [len(w) for w in words]

print(lengths)   # [6, 5, 9, 6]
````

````python
# Перевод всех строк в верхний регистр
cities = ['москва', 'париж', 'рим']
upper_cities = [city.upper() for city in cities]

print(upper_cities)   # ['МОСКВА', 'ПАРИЖ', 'РИМ']
````

````python
# Преобразование строки в список символов
chars = [c for c in 'Python']

print(chars)   # ['P', 'y', 't', 'h', 'o', 'n']
````

### Включение с условием (фильтрация)

В конец включения можно добавить условие `if` — тогда в новый список попадут только те элементы, для которых условие истинно.

__Синтаксис__
```python
[выражение for переменная in итерируемый_объект if условие]
```

````python
# Сравнение: оставить только чётные числа

# Обычный цикл
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)
print(evens)   # [0, 2, 4, 6, 8]
````

````python
# Списковое включение с условием
evens = [x for x in range(10) if x % 2 == 0]
print(evens)   # [0, 2, 4, 6, 8]
````

````python
# Квадраты только чётных чисел
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(even_squares)   # [0, 4, 16, 36, 64]
````

````python
# Оставить только короткие слова (длиной меньше 6 символов)
words = ['Python', 'МГИМО', 'включение', 'список', 'код']
short_words = [w for w in words if len(w) < 6]

print(short_words)   # ['МГИМО', 'код']
````

### Включение с условным выражением `if ... else`

Внутри включения можно использовать и **условное выражение** `if ... else` — то самое, что в обычном коде записывается как `a if условие else b`. Оно не фильтрует элементы, а **преобразует** их: выбирает одно из двух значений в зависимости от условия.

<div class="alert alert-warning">

**Не путайте два разных `if`:**

* `[x for x in ... if условие]` — `if` **в конце** работает как **фильтр** (пропускает или не пропускает элемент).
* `[x if условие else y for x in ...]` — `if ... else` **в начале** работает как **выбор значения** (элемент всегда попадает в список, но в одном из двух видов).

</div>

````python
# Для каждого числа от 0 до 9 записываем 'чётное' или 'нечётное'
labels = ['чётное' if x % 2 == 0 else 'нечётное' for x in range(10)]

print(labels)
# ['чётное', 'нечётное', 'чётное', 'нечётное', 'чётное',
#  'нечётное', 'чётное', 'нечётное', 'чётное', 'нечётное']
````

````python
# Заменим отрицательные числа на 0, а неотрицательные оставим как есть
numbers = [5, -3, 7, -1, 0, -8, 4]
clipped = [x if x >= 0 else 0 for x in numbers]

print(clipped)   # [5, 0, 7, 0, 0, 0, 4]
````

Можно комбинировать обе формы `if` в одном включении: сначала `if ... else` преобразует значения, а финальный `if` в конце — фильтрует.

````python
# Для чётных чисел от 0 до 9 записываем квадрат, для нечётных — куб
result = [x ** 2 if x % 2 == 0 else x ** 3 for x in range(10)]

print(result)   # [0, 1, 4, 27, 16, 125, 36, 343, 64, 729]
````

### Вложенные циклы во включении

Во включении можно использовать несколько `for` подряд — это работает как вложенные циклы. Порядок такой же, как если бы вы написали их обычным способом: первый `for` — внешний, второй — внутренний.

__Синтаксис__
```python
[выражение for x in ... for y in ...]
```

````python
# Все пары (цвет, объект) — сравните с примером из темы про циклы
colors = ["красный", "синий", "зелёный"]
objects = ["мяч", "кубик"]

pairs = [(color, obj) for color in colors for obj in objects]

print(pairs)
# [('красный', 'мяч'), ('красный', 'кубик'),
#  ('синий', 'мяч'),  ('синий', 'кубик'),
#  ('зелёный', 'мяч'), ('зелёный', 'кубик')]
````

````python
# Развернуть матрицу в одномерный список
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flat = [element for row in matrix for element in row]

print(flat)   # [1, 2, 3, 4, 5, 6, 7, 8, 9]
````

__Объяснение:__ Порядок `for` читается слева направо ровно так же, как при обычной вложенности:
```python
for row in matrix:         # внешний цикл
    for element in row:    # внутренний цикл
        flat.append(element)
```

## 3. Множественное включение (set comprehension)

Множественное включение устроено **точно так же**, как списковое, только используются **фигурные** скобки `{ }` вместо квадратных. Результатом будет множество — а значит, все дубликаты автоматически удалятся.

__Синтаксис__
```python
{выражение for переменная in итерируемый_объект}
```

````python
# Уникальные длины слов
words = ['Python', 'МГИМО', 'включение', 'список', 'код', 'набор']
unique_lengths = {len(w) for w in words}

print(unique_lengths)   # {3, 5, 6, 9}  — порядок может быть любым
````

````python
# Уникальные символы строки в нижнем регистре
text = "Programming"
letters = {ch.lower() for ch in text}

print(letters)   # {'p', 'r', 'o', 'g', 'a', 'm', 'i', 'n'}
````

````python
# Можно использовать фильтр if — ровно так же, как в списковом включении
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
unique_even = {x for x in numbers if x % 2 == 0}

print(unique_even)   # {2, 4}
````

## 4. Словарное включение (dict comprehension)

Словарное включение создаёт словарь: для каждого элемента задаются сразу **ключ** и **значение**, разделённые двоеточием.

__Синтаксис__
```python
{ключ: значение for переменная in итерируемый_объект}
```

````python
# Словарь "число → его квадрат"
squares = {x: x ** 2 for x in range(6)}

print(squares)   # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
````

````python
# Словарь "слово → его длина"
words = ['Python', 'МГИМО', 'включение']
lengths = {w: len(w) for w in words}

print(lengths)   # {'Python': 6, 'МГИМО': 5, 'включение': 9}
````

### Параллельный перебор двух коллекций с `zip()`

Функция `zip()` попарно соединяет элементы нескольких последовательностей. Вместе со словарным включением это удобный способ «склеить» два списка в словарь.

````python
students = ['Анна', 'Борис', 'Вера']
grades   = [5, 4, 5]

gradebook = {name: grade for name, grade in zip(students, grades)}

print(gradebook)   # {'Анна': 5, 'Борис': 4, 'Вера': 5}
````

### Фильтрация в словарном включении

Условие `if` работает ровно так же, как в списковом включении.

````python
# Оставим в словаре только отличников
grades = {'Анна': 5, 'Борис': 4, 'Вера': 5, 'Глеб': 3}

top_students = {name: grade for name, grade in grades.items() if grade == 5}

print(top_students)   # {'Анна': 5, 'Вера': 5}
````

````python
# Поменяем местами ключи и значения существующего словаря
original = {'a': 1, 'b': 2, 'c': 3}
swapped  = {value: key for key, value in original.items()}

print(swapped)   # {1: 'a', 2: 'b', 3: 'c'}
````

<div class="alert alert-warning">

**Обратите внимание:** когда мы перебираем словарь, сам по себе цикл `for key in dct` даёт только ключи. Чтобы получить одновременно ключ и значение, нужно использовать `dct.items()` — вы уже встречали этот метод в теме про словари.

</div>

## 5. Сводная таблица всех трёх включений

| Что создаём | Скобки | Пример | Результат |
|---|---|---|---|
| Список | `[ ]` | `[x ** 2 for x in range(4)]` | `[0, 1, 4, 9]` |
| Множество | `{ }` | `{x % 3 for x in range(10)}` | `{0, 1, 2}` |
| Словарь | `{ключ: значение}` | `{x: x ** 2 for x in range(4)}` | `{0: 0, 1: 1, 2: 4, 3: 9}` |

<div class="alert alert-warning">

**Про фигурные скобки:** и множество, и словарь используют `{ }`, а отличить их Python может только по наличию двоеточия. `{x for x in range(3)}` — это множество, а `{x: x for x in range(3)}` — словарь. Пустые же `{}` создают **пустой словарь**, а не множество; для пустого множества нужно использовать `set()`.

</div>

## 6. Когда использовать включения, а когда — обычный цикл

Включения — мощный инструмент, но далеко не всегда лучший выбор. Хорошее правило: **включение должно читаться легче, чем эквивалентный цикл**. Если нет — пишите цикл.

__Включение уместно, когда:__
- нужно создать новую коллекцию из существующей;
- логика преобразования/фильтрации простая и помещается в одну строку без потери читаемости.

__От включения лучше отказаться, когда:__
- в теле много условий и преобразований — получится нечитаемая «простыня»;
- внутри нужен вызов `print()`, ввод данных или другие *побочные действия* — в этом случае цикл семантически точнее;
- требуются операторы `break` или `continue` — во включениях их использовать нельзя;
- коллекция не нужна, а нужно просто что-то сделать N раз.

<div class="alert alert-success">

**Практический совет:** если вам приходится перечитывать собственное включение дважды, чтобы понять, что оно делает, — перепишите его обычным циклом. Короткий код не всегда хороший код; хороший код — это читаемый код.

</div>

### Пример: когда лучше обычный цикл

Представьте, что нужно по списку оценок вывести на экран сообщения о каждом студенте и попутно посчитать средний балл. Включение здесь — плохой выбор: нет новой коллекции, есть побочный вывод, считается агрегат.

````python
# Хорошо: обычный цикл с понятной логикой
grades = {'Анна': 5, 'Борис': 4, 'Вера': 5, 'Глеб': 3}

total = 0
for name, grade in grades.items():
    print(f"{name}: {grade}")
    total += grade

average = total / len(grades)
print(f"Средний балл: {average}")
````
